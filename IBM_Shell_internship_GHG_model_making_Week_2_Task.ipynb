# Supply Chain Emissions Modeling Using Industry and Commodity Data (2010–2016)

Problem Statement:

You have annual supply chain emission data from 2010–2016 categorized into industries and commodities. The goal is to develop a regression model that can predict the Supply Chain Emission Factors with Margins based on descriptive and quality metrics (substance, unit, reliability, temporal/geographical/technological/data collection correlations, etc.).

# Step 1: Import Required Libraries

```python

import pandas as pd # data analysis
import numpy as np  # for numerical data
import seaborn as sns # data visualization advanced
import matplotlib.pyplot as plt # basic
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler # Feature Scaling
from sklearn.ensemble import RandomForestRegressor  # numeric and categoric data together
from sklearn.metrics import mean_squared_error, r2_score
import joblib
import os
```

# Step 2: Load Dataset

```python
excel_file = '/content/SupplyChainEmissionFactorsforUSIndustriesCommodities (1).xlsx'  # Replace with actual path
years = range(2010, 2017)

# df = pd.read_excel(r"path_data.xlsx")
```

```python
df_2 = pd.read_excel(excel_file, sheet_name=f'{years[0]}_Detail_Industry')
df_2.head()
```

```python
all_data = []

for year in years:
    try:
        df_com = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Commodity')
        df_ind = pd.read_excel(excel_file, sheet_name=f'{year}_Detail_Industry')

        df_com['Source'] = 'Commodity'
        df_ind['Source'] = 'Industry'
        df_com['Year'] = df_ind['Year'] = year

        df_com.columns = df_com.columns.str.strip()
        df_ind.columns = df_ind.columns.str.strip()

        df_com.rename(columns={
            'Commodity Code': 'Code',
            'Commodity Name': 'Name'
        }, inplace=True)

        df_ind.rename(columns={
            'Industry Code': 'Code',
            'Industry Name': 'Name'
        }, inplace=True)

        all_data.append(pd.concat([df_com, df_ind], ignore_index=True))

    except Exception as e:
        print(f"Error processing year {year}: {e}")
```

```python
df = pd.concat(all_data, ignore_index=True)
df.head()
```

## Normalize features

```python
# Normalize features
scaler = StandardScaler()
# Convert 'Margins of Supply Chain Emission Factors' to numeric, coercing errors and filling NaN
X['Margins of Supply Chain Emission Factors'] = pd.to_numeric(X['Margins of Supply Chain Emission Factors'], errors='coerce').fillna(0)
X_scaled = scaler.fit_transform(X)
```

#### Divide the data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Select the model for training

```python
model = RandomForestRegressor(random_state=42)
```

### Training

```python
model.fit(X_train, y_train)
```

### Prediction anf Evaluation

```python
y_pred = model.predict(X_test)
# r2_sccore approx 1 gives us perfecr model
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'RMSE: {rmse}')
print(f'R² Score: {r2}')
```

```python
from sklearn.linear_model import LinearRegression

# Select the model for training
linear_model = LinearRegression()

# Training
linear_model.fit(X_train, y_train)

# Prediction
y_pred_linear = linear_model.predict(X_test)

# Evaluation
mse_linear = mean_squared_error(y_test, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y_test, y_pred_linear)

print(f'Linear Regression RMSE: {rmse_linear}')
print(f'Linear Regression R² Score: {r2_linear}')
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Select the model for training
knn_model = KNeighborsRegressor()

# Training
knn_model.fit(X_train, y_train)

# Prediction
y_pred_knn = knn_model.predict(X_test)

# Evaluation
mse_knn = mean_squared_error(y_test, y_pred_knn)
rmse_knn = np.sqrt(mse_knn)
r2_knn = r2_score(y_test, y_pred_knn)

print(f'KNN Regressor RMSE: {rmse_knn}')
print(f'KNN Regressor R² Score: {r2_knn}')
```

```python
from sklearn.svm import SVR

# Select the model for training
svr_model = SVR()

# Training
svr_model.fit(X_train, y_train)

# Prediction
y_pred_svr = svr_model.predict(X_test)

# Evaluation
mse_svr = mean_squared_error(y_test, y_pred_svr)
rmse_svr = np.sqrt(mse_svr)
r2_svr = r2_score(y_test, y_pred_svr)

print(f'Support Vector Regressor RMSE: {rmse_svr}')
print(f'Support Vector Regressor R² Score: {r2_svr}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Select the model for training
gbr_model = GradientBoostingRegressor(random_state=42)

# Training
gbr_model.fit(X_train, y_train)

# Prediction
y_pred_gbr = gbr_model.predict(X_test)

# Evaluation
mse_gbr = mean_squared_error(y_test, y_pred_gbr)
rmse_gbr = np.sqrt(mse_gbr)
r2_gbr = r2_score(y_test, y_pred_gbr)

print(f'Gradient Boosting Regressor RMSE: {rmse_gbr}')
print(f'Gradient Boosting Regressor R² Score: {r2_gbr}')
```

### Hyperparameter Tuning

```python
# When the model performs very well on the training data we say that it is overfit

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
```

### Use best parameters for prediction

```python
y_pred_best = best_model.predict(X_test)

print("Final RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_best)))
print("Final R²:", r2_score(y_test, y_pred_best))
```

# Task
Visualize the decision regions for each trained model (Random Forest, Linear Regression, KNN, SVR, Gradient Boosting) after applying dimensionality reduction to the feature data.

## Dimensionality reduction

### Subtask:
Apply a dimensionality reduction technique (like PCA) to reduce the feature space to 2 or 3 dimensions.


**Reasoning**:
Apply PCA to reduce the dimensionality of the features to 2 components for visualization.



```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

## Model training on reduced data

### Subtask:
Retrain each of the models (Random Forest, Linear Regression, KNN, SVR, Gradient Boosting) using the dimensionality-reduced training data.


**Reasoning**:
Train the RandomForestRegressor model using the dimensionality-reduced training data.



```python
rf_pca = RandomForestRegressor(random_state=42)
rf_pca.fit(X_train_pca, y_train)
```

**Reasoning**:
Train the LinearRegression, KNeighborsRegressor, SVR, and GradientBoostingRegressor models using the dimensionality-reduced training data.



```python
lr_pca = LinearRegression()
lr_pca.fit(X_train_pca, y_train)

knn_pca = KNeighborsRegressor()
knn_pca.fit(X_train_pca, y_train)

svr_pca = SVR()
svr_pca.fit(X_train_pca, y_train)

gbr_pca = GradientBoostingRegressor(random_state=42)
gbr_pca.fit(X_train_pca, y_train)
```

## Create a meshgrid

### Subtask:
Generate a meshgrid of points covering the range of the reduced feature space.


**Reasoning**:
Generate a meshgrid covering the range of the two principal components to prepare for visualizing decision regions.



```python
x_min, x_max = X_train_pca[:, 0].min() - 1, X_train_pca[:, 0].max() + 1
y_min, y_max = X_train_pca[:, 1].min() - 1, X_train_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
X_grid = np.c_[xx.ravel(), yy.ravel()]
```

## Predict on meshgrid

### Subtask:
Use each retrained model to predict the target variable for each point in the meshgrid.


**Reasoning**:
Use each retrained model to predict the target variable for each point in the meshgrid.



```python
Z_rf = rf_pca.predict(X_grid)
Z_lr = lr_pca.predict(X_grid)
Z_knn = knn_pca.predict(X_grid)
Z_svr = svr_pca.predict(X_grid)
Z_gbr = gbr_pca.predict(X_grid)
```

```python
# Evaluate each model on the PCA-transformed test data
models_pca = {
    'Random Forest (PCA)': rf_pca,
    'Linear Regression (PCA)': lr_pca,
    'KNN Regressor (PCA)': knn_pca,
    'Support Vector Regressor (PCA)': svr_pca,
    'Gradient Boosting Regressor (PCA)': gbr_pca
}

for name, model in models_pca.items():
    y_pred_pca = model.predict(X_test_pca)
    mse_pca = mean_squared_error(y_test, y_pred_pca)
    rmse_pca = np.sqrt(mse_pca)
    r2_pca = r2_score(y_test, y_pred_pca)

    print(f'{name} Evaluation:')
    print(f'  RMSE: {rmse_pca}')
    print(f'  R² Score: {r2_pca}')
    print("-" * 20)
```

## Visualize decision regions

### Subtask:
Plot the meshgrid points, colored by their predicted class (or a continuous color scale for regression), and overlay the original data points.


**Reasoning**:
Plot the decision regions and original data points for each model.



```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

models = {
    'Random Forest': (Z_rf, rf_pca),
    'Linear Regression': (Z_lr, lr_pca),
    'KNN Regressor': (Z_knn, knn_pca),
    'Support Vector Regressor': (Z_svr, svr_pca),
    'Gradient Boosting Regressor': (Z_gbr, gbr_pca)
}

# Determine a consistent color scale across all plots
all_predictions = np.concatenate([Z_rf, Z_lr, Z_knn, Z_svr, Z_gbr])
vmin = min(all_predictions.min(), y_test.min())
vmax = max(all_predictions.max(), y_test.max())

for i, (title, (Z, model)) in enumerate(models.items()):
    ax = axes[i]
    Z = Z.reshape(xx.shape)
    contour = ax.contourf(xx, yy, Z, cmap='viridis', vmin=vmin, vmax=vmax)
    scatter = ax.scatter(X_test_pca[:, 0], X_test_pca[:, 1], c=y_test, cmap='viridis', edgecolors='k', s=20, vmin=vmin, vmax=vmax)
    ax.set_title(f'Decision Regions ({title})')
    ax.set_xlabel('Principal Component 1')
    ax.set_ylabel('Principal Component 2')
    fig.colorbar(contour, ax=ax, label='Predicted Emission Factor')

# Remove the empty subplot if there is one
if len(models) < len(axes):
    fig.delaxes(axes[len(models)])


plt.tight_layout()
plt.show()
```
